```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("data/country dataset.csv");
```

```python
df.columns
```

```python
plt.scatter(df['latitude'], df['longitude'])
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
features = df.drop(['country'], axis=1)
```

```python
model = KMeans(5)
model.fit(features)
```

```python
clusters = model.predict(features)
```

```python
df2 = features
df2['clusters'] = clusters
```

```python
df2
```

```python
plt.scatter(df2['latitude'], df2['longitude'], c=df2['clusters'], cmap='rainbow')
```

```python
ll = []
for i in range(1, 10):
    kmeans = KMeans(i)
    kmeans.fit(features)
    ins = kmeans.inertia_
    ll.append(ins)

n_cluster = range(1, 10)
plt.plot(n_cluster, ll)
plt.xlabel('n clusters')
plt.ylabel('Insersia')
```
